```python
gdp_df = pd.read_csv("countrygdp.csv")


gdp_df.head()
```

```python

year_cols = [col for col in gdp_df.columns if col.strip().isdigit()]


gdp_long = gdp_df.melt(
    id_vars=["Country Name", "Country Code"], 
    value_vars=year_cols,                      
    var_name="Year",
    value_name="GDP"
)

#convert year to integer
gdp_long["Year"] = gdp_long["Year"].astype(int)

#for merging
gdp_long = gdp_long.rename(columns={"Country Code": "Country"})

#importer GDP
copy_df = copy_df.merge(
    gdp_long.rename(columns={"Country": "Importer", "GDP": "Importer_GDP"})[["Importer", "Year", "Importer_GDP"]],
    how="left",
    on=["Importer", "Year"]
)

#exporter GDP
copy_df = copy_df.merge(
    gdp_long.rename(columns={"Country": "Exporter", "GDP": "Exporter_GDP"})[["Exporter", "Year", "Exporter_GDP"]],
    how="left",
    on=["Exporter", "Year"]
)
```

```python
copy_df.isna().sum()
```

```python
#year is numeric
copy_df['Year'] = pd.to_numeric(copy_df['Year'], errors='coerce')

#group by taxon and year
species_yearly_trade = (
    copy_df.groupby(['Taxon', 'Year'])['Quantity']
    .sum()
    .reset_index()
    .sort_values(['Taxon', 'Year'])
)
```

```python
species_yearly_trade['TradeChange'] = species_yearly_trade.groupby('Taxon')['Quantity'].pct_change() * 100
```

```python
#total precent change year to year
species_trend = species_yearly_trade.groupby('Taxon').agg(
    FirstYear=('Year', 'min'),
    LastYear=('Year', 'max'),
    StartQty=('Quantity', lambda x: x.iloc[0]),
    EndQty=('Quantity', lambda x: x.iloc[-1])
).reset_index()

species_trend['OverallChangePercent'] = ((species_trend['EndQty'] - species_trend['StartQty']) / species_trend['StartQty']) * 100
```

```python
risk_threshold = 5000  # you can tune this later
species_trend['AtRisk'] = species_trend['OverallChangePercent'] > risk_threshold
```

```python
labeled_data = copy_df.merge(species_trend[['Taxon', 'AtRisk']], on='Taxon', how='left')
```

```python
copy_df = copy_df.drop(columns=[col for col in copy_df.columns if col.startswith('AtRisk')], errors='ignore')
```

```python
#species-year trends
copy_df['Year'] = pd.to_numeric(copy_df['Year'], errors='coerce')

species_yearly_trade = (
    copy_df.groupby(['Taxon', 'Year'])['Quantity']
    .sum()
    .reset_index()
    .sort_values(['Taxon', 'Year'])
)

species_yearly_trade['TradeChange'] = (
    species_yearly_trade.groupby('Taxon')['Quantity']
    .pct_change() * 100
)

#risk threshold
risk_threshold = 25
species_yearly_trade['AtRisk'] = species_yearly_trade['TradeChange'] > risk_threshold

#merge
copy_df = copy_df.merge(
    species_yearly_trade[['Taxon', 'Year', 'AtRisk']],
    on=['Taxon', 'Year'],
    how='left'
)

#fixes if merges its no at risk
if 'AtRisk' not in copy_df.columns:
    copy_df['AtRisk'] = False
else:
    copy_df['AtRisk'] = copy_df['AtRisk'].fillna(False)

#labeled dataset
labeled_data = copy_df
```

```python
#select features
selected_features = [
    'Quantity', 'Importer_GDP', 'Exporter_GDP'
    , 'Importer', 'Exporter', 'Purpose', 'Source'
]

#drop rows where no atrisk
df_model = labeled_data.dropna(subset=['AtRisk'])

# split into x and y features
X = df_model[selected_features]
y = df_model['AtRisk']
```

```python
from sklearn.model_selection import train_test_split

#split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#separate features
numerical_features = ['Quantity', 'Importer_GDP', 'Exporter_GDP']
categorical_features = [ 'Importer', 'Exporter','Purpose', 'Source']

#pipelines
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

#preprocessor
preprocessor = ColumnTransformer([
    ('num', num_pipeline, numerical_features),
    ('cat', cat_pipeline, categorical_features)
])
```

```python
from sklearn.tree import DecisionTreeClassifier

#define classifier, adjust max depth
tree_clf = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(max_depth=4, random_state=42))
])

#fit model
tree_clf.fit(X_train, y_train)
```

```python
#one-hot encoded features
ohe = tree_clf.named_steps['preprocessor'].named_transformers_['cat'].named_steps['encoder']
cat_columns = ohe.get_feature_names_out(categorical_features)

#combine with numerical features
feature_names = numerical_features + list(cat_columns)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

#feature list
feature_cols_clf = [
    'Term', 'Unit', 'Importer', 'Exporter', 'Origin',
    'Purpose', 'Source', 'Year', 'Importer_GDP', 'Exporter_GDP'
]

#sample data
sampled_df_clf = labeled_data.sample(n=1_000_000, random_state=42)
X_clf = sampled_df_clf[feature_cols_clf]
y_clf = sampled_df_clf['AtRisk']

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

#preprocessor and features
num_cols = ['Year', 'Importer_GDP', 'Exporter_GDP']
cat_cols = ['Term', 'Unit', 'Importer', 'Exporter', 'Origin', 'Purpose', 'Source']

preprocessor_clf = ColumnTransformer(transformers=[
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), num_cols),
    
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), cat_cols)
])

#modeling with set tree depth of 5
bagging_clf_pipe = Pipeline([
    ('preprocessor', preprocessor_clf),
    ('classifier', BaggingClassifier(
        base_estimator=DecisionTreeClassifier(max_depth=5, class_weight='balanced'),
        n_estimators=50,
        n_jobs=-1,
        random_state=42
    ))
])

#train and evaluate
bagging_clf_pipe.fit(X_train_clf, y_train_clf)
y_pred_bag_clf = bagging_clf_pipe.predict(X_test_clf)

print(" Classification Report (Bagging Classifier):")
print(classification_report(y_test_clf, y_pred_bag_clf))

```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report
from sklearn.inspection import permutation_importance
import pandas as pd
import numpy as np


#features

feature_cols_clf = [
    'Term', 'Unit', 'Importer', 'Exporter', 'Origin',
    'Purpose', 'Source', 'Year', 'Importer_GDP', 'Exporter_GDP'
]
num_cols = ['Year', 'Importer_GDP', 'Exporter_GDP']
cat_cols = list(set(feature_cols_clf) - set(num_cols))


#sample smaller set of data

sampled_df_clf = labeled_data.sample(n=200000, random_state=42)
X_clf = sampled_df_clf[feature_cols_clf]
y_clf = sampled_df_clf['AtRisk']

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, stratify=y_clf, random_state=42
)


#preprocess

preprocessor_clf = ColumnTransformer(transformers=[
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), num_cols),
    
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore', sparse=False))
    ]), cat_cols)
])


#create pipeline

bagging_clf_pipe = Pipeline([
    ('preprocessor', preprocessor_clf),
    ('classifier', BaggingClassifier(
        base_estimator=DecisionTreeClassifier(max_depth=5, class_weight='balanced'),
        n_estimators=50,
        n_jobs=-1,
        random_state=42
    ))
])

#fit model on train and test set
bagging_clf_pipe.fit(X_train_clf, y_train_clf)

print(" BaggingClassifier Results (Test Set):")
y_pred_test = bagging_clf_pipe.predict(X_test_clf)
print(classification_report(y_test_clf, y_pred_test))

print("\n BaggingClassifier Results (Train Set):")
y_pred_train = bagging_clf_pipe.predict(X_train_clf)
print(classification_report(y_train_clf, y_pred_train))

#cv on f1-macro
print("3-Fold Stratified CV")
strat_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
cv_scores = cross_val_score(
    bagging_clf_pipe, X_clf, y_clf,
    scoring='f1_macro', cv=strat_kfold, n_jobs=-1
)
print(" CV F1 (macro):", cv_scores)
print("Mean:", cv_scores.mean(), "Std:", cv_scores.std())


```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

#features
feature_cols = [
    'Term', 'Unit', 'Importer', 'Exporter', 'Origin',
    'Purpose', 'Source', 'Year', 'Importer_GDP', 'Exporter_GDP'
]

#sample 1million rows
sampled_df = labeled_data.sample(n=1_000_000, random_state=42)
X = sampled_df[feature_cols]
y = sampled_df['AtRisk']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

#preprocesses
numeric_cols = ['Year', 'Importer_GDP', 'Exporter_GDP']
categorical_cols = [col for col in feature_cols if col not in numeric_cols]

preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numeric_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), categorical_cols)
])

#mlp pipeline
mlp_clf = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42))
])

#test on smaller set first
mlp_clf.fit(X_train, y_train)
y_pred_mlp = mlp_clf.predict(X_test)

print(" MLPClassifier Results")
print(classification_report(y_test, y_pred_mlp))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import StratifiedKFold

features = ['Term', 'Unit', 'Importer', 'Exporter', 'Origin', 'Purpose', 'Source', 'Year', 'Importer_GDP', 'Exporter_GDP']
target = 'AtRisk'

X = labeled_data[features]
y = labeled_data[target]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)


numeric_features = ['Year', 'Importer_GDP', 'Exporter_GDP']
categorical_features = [col for col in features if col not in numeric_features]

preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numeric_features),
    
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), categorical_features)
])

mlp_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', MLPClassifier(hidden_layer_sizes=(100,), max_iter=200, random_state=42))
])

#fit model and test cv
mlp_pipe.fit(X_train, y_train)
y_pred = mlp_pipe.predict(X_test)
print(" MLPClassifier Results (Test Set)")
print(classification_report(y_test, y_pred))

y_pred_train = mlp_pipe.predict(X_train)
print(" MLPClassifier Results (Training Set)")
print(classification_report(y_train, y_pred_train))



print(" 5-fold Stratified Cross-Validation")
strat_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    mlp_pipe, X, y,
    scoring='f1_macro', cv=strat_kfold, n_jobs=-1
)

print(" Cross-validated F1 (macro):")
print(f"Scores: {cv_scores}")
print(f"Mean: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")

```
